# Q2 What is sampling in statistics? Explain the differences between random and stratified sampling.

Sampling is the process of selecting a subset (sample) from a larger group (population) to study and draw conclusions about the entire population.
It is used because studying the whole population is often time-consuming, costly, or impractical.
1. Random Sampling

Meaning:
In random sampling, every member of the population has an equal and independent chance of being selected.

How it works:

Use random numbers

Lottery method

Computer-generated random selection

Example:
A company has 1,000 employees. You randomly select 100 employee IDs using Excel‚Äôs RAND() function.

Advantages:

Simple to use

Reduces selection bias

Easy to implement

Disadvantages:

May not represent all subgroups well

Important groups can be under-represented

2. Stratified Sampling

Meaning:
In stratified sampling, the population is first divided into subgroups (strata) based on a common characteristic, and then random samples are taken from each group.

Common strata examples:

Department

Gender

Age group

Experience level

Example:
A company has:

60% Developers

25% Testers

15% Managers

To select 100 employees:

60 Developers

25 Testers

15 Managers

are randomly chosen from each group.

Advantages:

Better representation of all groups

More accurate results

Lower sampling error

Disadvantages:

More complex to design

Requires detailed population information

Why Are These Measures of Central Tendency Important?

# Q4  Explain skewness and kurtosis. What does a positive skew imply about the data? 

Skewness

Skewness measures the asymmetry of a data distribution around its mean‚Äîi.e., whether the data leans more to the left or right.

Types of Skewness

Zero skew (Symmetrical): Left and right sides are mirror images

Positive skew (Right-skewed): Long tail on the right side

Negative skew (Left-skewed): Long tail on the left side

Relationship (rule of thumb):

Positive skew: Mean > Median > Mode

Negative skew: Mean < Median < Mode

Example (Income data):
Most people earn moderate incomes, but a few earn very high incomes ‚Üí positive skew

Kurtosis

Kurtosis measures the tailedness or peakedness of a distribution‚Äîhow heavy or light the tails are compared to a normal distribution.

Types of Kurtosis

Mesokurtic: Normal distribution (moderate peak, moderate tails)

Leptokurtic: High peak, heavy tails (more extreme values/outliers)

Platykurtic: Flat peak, light tails (fewer outliers)

Note:
Kurtosis focuses more on outliers than on the height of the peak.

What Does a Positive Skew Imply About the Data?

A positive skew means:

Most values are concentrated on the lower side

A few very high values stretch the right tail

The mean is pulled upward by extreme high values

Median is a better measure of central tendency than mean

# Q6 Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python: list_x = [10, 20, 30, 40, 50] list_y = [15, 25, 35, 45, 60] 

```python


list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert to NumPy arrays
x = np.array(list_x)
y = np.array(list_y)

# Covariance (sample covariance)
cov_matrix = np.cov(x, y)
covariance = cov_matrix[0, 1]

# Correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

# Q7  Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result: data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35] 

```python
import numpy as np
import matplotlib.pyplot as plt

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Convert to NumPy array
arr = np.array(data)

# Quartiles
Q1 = np.percentile(arr, 25)
Q3 = np.percentile(arr, 75)
IQR = Q3 - Q1

# Outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = arr[(arr < lower_bound) | (arr > upper_bound)]

# Boxplot
plt.boxplot(arr)
plt.title("Boxplot of Data")
plt.ylabel("Values")
plt.show()

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

# Q8  You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales. ‚óè Explain how you would use covariance and correlation to explore this relationship. ‚óè Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500]   daily_sales = [2200, 2450, 2750, 3200, 4000] 

1Ô∏è‚É£ Covariance ‚Äì Direction of Relationship

Covariance tells us whether both variables move together.

Positive covariance ‚Üí higher ad spend tends to increase sales

Negative covariance ‚Üí higher ad spend tends to decrease sales

Limitation: The value depends on scale, so it‚Äôs hard to interpret strength.

2Ô∏è‚É£ Correlation ‚Äì Strength of Relationship

Correlation standardizes covariance into a value between -1 and +1.

Helps answer:

How strong is the relationship?

Is it positive or negative?

Interpretation:

+1 ‚Üí Perfect positive relationship

0 ‚Üí No relationship

-1 ‚Üí Perfect negative relationship


Explanation of the Result

The correlation value is close to +1, indicating a strong positive relationship.

This suggests that increasing advertising spend is strongly associated with higher daily sales.

While correlation does not imply causation, this result supports further analysis (e.g., regression).

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to NumPy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Correlation coefficient
correlation = np.corrcoef(x, y)[0, 1]

print("Correlation between advertising spend and daily sales:", correlation)
```

# Q9 Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product. ‚óè Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you‚Äôd use. ‚óè Write Python code to create a histogram using Matplotlib for the survey data: survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7] 

1Ô∏è‚É£ Mean

Shows the average satisfaction level.

2Ô∏è‚É£ Median

Indicates the typical score, especially useful if data is skewed.
