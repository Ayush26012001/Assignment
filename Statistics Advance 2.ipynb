# Question 1: What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make decisions or inferences about a population parameter based on sample data. It helps determine whether there is enough evidence to reject a stated assumption (hypothesis) about the population.

# Q2 What is the null hypothesis, and how does it differ from the alternative hypothesis?
- Null Hypothesis (H₀): It states that there is no effect, no difference, or no relationship in the population.

- Alternative Hypothesis (H₁ or Hₐ): It states that there is an effect, a difference, or a relationship.

- Difference:
- The null hypothesis assumes the status quo, while the alternative hypothesis represents what we want to prove.


# Q3 Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
- The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. Common values are 0.05 or 0.01.
- If the p-value ≤ α, we reject the null hypothesis; otherwise, we fail to reject it.


# Q4 What are Type I and Type II errors? Give examples of each.
- Type I Error: Rejecting a true null hypothesis.
- Example: Declaring a medicine effective when it is not.

- Type II Error: Failing to reject a false null hypothesis.
- Example: Declaring a medicine ineffective when it actually works.


# Q5 What is the difference between a Z-test and a T-test? Explain when to use each.

#### Z-Test
- Population standard deviation is known	
- Large sample size (n ≥ 30)	
- Uses normal distribution

#### T-Test
- Population standard deviation is unknown
- Small sample size (n < 30)
- Uses t-distribution

# Q6 Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import random
import matplotlib.pyplot as plt

# Generate binomial data
data = [sum(1 for _ in range(10) if random.random() < 0.5) for _ in range(1000)]

# Plot histogram
plt.hist(data, bins=11)
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()
```

# Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Interpret the results.

```python
import numpy as np
from math import sqrt

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

population_mean = 50
population_std = 1
n = len(sample_data)

sample_mean = np.mean(sample_data)

z = (sample_mean - population_mean) / (population_std / sqrt(n))
print("Z-value:", z)
```

# Q8 Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(50, 5, 100)
mean = np.mean(data)
std = np.std(data)
n = len(data)

ci = norm.interval(0.95, mean, std / np.sqrt(n))
print("95% Confidence Interval:", ci)

plt.hist(data, bins=15)
plt.title("Normal Distribution Data")
plt.show()
```

# Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain Z-scores.

```python
import numpy as np
import matplotlib.pyplot as plt

def z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return [(x - mean) / std for x in data]

data = [10, 12, 15, 18, 20, 22, 25]
z_data = z_scores(data)

plt.hist(z_data, bins=7)
plt.title("Z-Score Histogram")
plt.show()
```
