

## **Question 1: What is a random variable in probability theory?**

### **Answer:**

A **random variable** is a numerical function that assigns a real number to each outcome of a random experiment.

**Example:**

* Tossing a coin ‚Üí Random variable = number of heads
* Rolling a die ‚Üí Random variable = value on the die

---

## **Question 2: What are the types of random variables?**

### **Answer:**

There are **two main types**:

1. **Discrete random variables** ‚Äì take countable values
   (e.g., number of customers, number of defects)
2. **Continuous random variables** ‚Äì take any value in a range
   (e.g., height, weight, time)

---

## **Question 4: What is a binomial distribution, and how is it used in probability?**

### **Answer:**

A **binomial distribution** models the number of successes in a fixed number of **independent trials**, where each trial has:

* Two outcomes (success/failure)
* Constant probability of success

**Example:**
Number of customers who purchase after seeing an ad.

---

## **Question 5: What is the standard normal distribution, and why is it important?**

### **Answer:**

The **standard normal distribution** is a normal distribution with:

* Mean = 0
* Standard deviation = 1

It is important because:

* Any normal distribution can be converted to it using **z-scores**
* Widely used in hypothesis testing and confidence intervals

---

## **Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

### **Answer:**

The **Central Limit Theorem (CLT)** states that:

> The sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the population‚Äôs distribution.

**Why it‚Äôs critical:**

* Enables confidence intervals and hypothesis testing
* Makes statistical inference possible for real-world data

---

## **Question 7: What is the significance of confidence intervals in statistical analysis?**

### **Answer:**

A **confidence interval** provides a **range of values** within which the true population parameter is likely to lie.

**Example:**
95% CI means we are 95% confident the true mean lies in that interval.

---

## **Question 8: What is the concept of expected value in a probability distribution?**

### **Answer:**

The **expected value** is the long-run average outcome of a random variable.

[
E(X) = \sum x \cdot P(x)
]

**Example:**
Expected revenue per customer in an e-commerce platform.




# Q9  Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute statistics
mean_value = np.mean(data)
std_value = np.std(data)

print("Mean:", mean_value)
print("Standard Deviation:", std_value)

# Histogram
plt.hist(data, bins=30)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Normal Distribution (Mean=50, Std=5)")
plt.show()
```

```python
import numpy as np
import math

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

n = len(daily_sales)
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)

# 95% confidence interval
z = 1.96
margin_of_error = z * (std_sales / math.sqrt(n))

lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", (lower_bound, upper_bound))
```
