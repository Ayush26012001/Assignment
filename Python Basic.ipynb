# Q4 Write a program that checks if a given word is a Python keyword.

```python
import keyword

word = input("Enter a word: ")

if keyword.iskeyword(word):
    print(f"✅ '{word}' is a Python keyword.")
else:
    print(f"❌ '{word}' is not a Python keyword.")
```

# Q5 Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

```python

my_list = [10, 20, 30]
my_tuple = (10, 20, 30)


print("Before change (list):", my_list)
my_list[1] = 99
print("After change (list):", my_list)


print("\nBefore change (tuple):", my_tuple)
try:
    my_tuple[1] = 99
except TypeError as e:
    print("Error:", e)
```

# Q6 Write a function to demonstrate the behavior of mutable and immutable arguments.

```python

def modify(x, y):
    x = x + 5          
    y.append(100)      
    print("Inside function:")
    print("x =", x)
    print("y =", y)


a = 10


b = [1, 2, 3]


modify(a, b)

print("\nOutside function:")
print("a =", a)
print("b =", b)
```

# Q7  Write a program that performs basic arithmetic operations on two user-input numbers

```python

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

add = num1 + num2
sub = num1 - num2
mul = num1 * num2


if num2 != 0:
    div = num1 / num2
else:
    div = "Undefined (cannot divide by zero)"


print("\n--- Results ---")
print("Addition:", add)
print("Subtraction:", sub)
print("Multiplication:", mul)
print("Division:", div)
```

# Q10  Write code to demonstrate type casting with list elements.


```python
numbers = ['10', '20', '30', '40']

int_list = [int(num) for num in numbers]
float_list = [float(num) for num in numbers]

print("Original list:", numbers)
print("After casting to integers:", int_list)
print("After casting to floats:", float_list)
```

# Q14  Write a program to reverse a string using a while loop.

```python
text = input("Enter a string: ")
reversed_text = ""
i = len(text)

while i > 0:
    reversed_text += text[i - 1]
    i -= 1

print("Reversed string:", reversed_text)
```

# Q15 Write a Python program to calculate the factorial of a number provided by the user using a while loop.

```python
num=5
fact=1
while num>0:
    fact=fact*num
    num=num-1
print(fact)
```
