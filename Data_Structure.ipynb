### 1. What are data structures, and why are they important?
Data structures are organized ways to store and manage data efficiently. They are important because they help optimize data access, modification, and storage depending on the task.

### 2. Difference between mutable and immutable data types with examples
Mutable types can be changed after creation (e.g., lists, dictionaries). Immutable types cannot be changed after creation (e.g., strings, tuples).

### 3. Differences between lists and tuples in Python
Lists are mutable, slower, and written with []. Tuples are immutable, faster, and written with ().

### 4. How dictionaries store data
Dictionaries store data as keyâ€“value pairs using a hash table internally.

### 5. Why might you use a set instead of a list in Python
Sets automatically remove duplicates and provide faster membership tests.

### 6. What is a string in Python, and how is it different from a list
A string is an immutable sequence of characters, whereas a list is mutable and can store different data types.

### 7. How tuples ensure data integrity
Tuples are immutable, so their values cannot be changed, ensuring data remains constant.

### 8. What is a hash table, and how does it relate to dictionaries
A hash table maps keys to values using a hashing function; dictionaries use hash tables internally.

### 9. Can lists contain different data types?
Yes, lists can contain mixed data types.

### 10. Explain why strings are immutable
Strings are immutable to ensure memory efficiency, security, and reliable hashing.

### 11. Advantages dictionaries offer over lists
Dictionaries provide fast lookups using keys and are better for structured data.

### 12. How sets handle duplicate values
Sets automatically remove duplicate values.

### 13. Scenario where a tuple is preferable over a list
When you want constant, unmodifiable data like coordinates or configuration values.

### 14. How the 'in' keyword works differently for lists and dictionaries
For lists, 'in' checks values; for dictionaries, 'in' checks only keys.

### 15. Can you modify elements of a tuple?
No, tuples are immutable, so elements cannot be changed.

### 16.What is a nested dictionary, and give an example of its use case
A dictionary containing another dictionary. Example: {'student': {'name': 'Ayu', 'age': 21}}.

### 17. Describe the time complexity of accessing elements in a dictionary
Average time complexity is O(1).

### 18. In what situations are lists preferred over dictionaries
When order matters or you only need a simple sequence of items.

### 19. Why dictionaries are considered unordered and how does that affect data retrieval
Because they store data based on hash values, not sequence order.

### 20. Difference between a list and a dictionary in terms of data retrieval
Lists retrieve items by index; dictionaries retrieve items by key.
