# 1. What is the difference between a function and a method in Python?
- Function: A function is a block of reusable code that is defined independently. It can be called by its name from anywhere in the program.
- Method: A method is a function that is associated with an object or a class. It is explicitly defined within a class definition and operates on the data of that object. You must call a method using an instance of the class (or the class itself, for class methods/static methods).

# 2. Explain the concept of function arguments and parameters in Python.
- Parameter: A parameter is the name used in the function definition to represent the value that will be passed into the function. They act as placeholders.
- Argument: An argument is the actual value that is passed to the function when it is called.

# Q5. What are iterators in Python and how do they differ from iterables?

-An iterable is any object that can be looped over.
-Examples: list, tuple, string, dictionary, set.

-An iterator is an object that produces values one at a time using:

-iter() → gets the iterator

-next() → gets the next value

-Difference:

-Iterable → can be looped over but does not maintain state.

-Iterator → maintains state and generates values on demand.

```python
# Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

# Check prime numbers from 1 to 200
for n in range(1, 201):
    if is_prime(n):
        print(n)
```

```python
#  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class Fibonacci:
    def __init__(self, n):
        self.n = n        # total terms
        self.count = 0    # current index
        self.a = 0        # first Fibonacci number
        self.b = 1        # second Fibonacci number

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # stop iteration

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b


# ----- USING THE ITERATOR -----
fib = Fibonacci(10)   # Generate first 10 Fibonacci numbers

for num in fib:
    print(num)
```

```python
# Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i


# --- Using the generator ---
for value in powers_of_two(10):   # yields 2^0 to 2^10
    print(value)
```

```python
# Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.rstrip("\n")   # remove newline
```

```python
for line in read_file_line_by_line("example.txt"):
    print(line)
```

```python
#  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
data = [(4, 10), (1, 3), (7, 5), (2, 1)]

sorted_list = sorted(data, key=lambda x: x[1])

print(sorted_list)
```

```python
#  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius = [0, 10, 20, 30, 40]

fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))

print(fahrenheit)
```

```python
# Create a Python program that uses filter() to remove all the vowels from a given string.
text = "Hello World, Python is Awesome!"

vowels = "aeiouAEIOU"

result = "".join(filter(lambda ch: ch not in vowels, text))

print(result)
```
